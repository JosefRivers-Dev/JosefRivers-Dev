```python
import pandas as pd

Data = "https://raw.githubusercontent.com/JosefRivers-Dev/Data-Science_Alura-Store_Oracle-LATAM/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
Data2 = "https://raw.githubusercontent.com/JosefRivers-Dev/Data-Science_Alura-Store_Oracle-LATAM/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
Data3 = "https://raw.githubusercontent.com/JosefRivers-Dev/Data-Science_Alura-Store_Oracle-LATAM/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
Data4 = "https://raw.githubusercontent.com/JosefRivers-Dev/Data-Science_Alura-Store_Oracle-LATAM/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(Data)
tienda2 = pd.read_csv(Data2)
tienda3 = pd.read_csv(Data3)
tienda4 = pd.read_csv(Data4)

tienda.head()
```

```python
tienda['Tienda'] = 'T1'
tienda2['Tienda'] = 'T2'
tienda3['Tienda'] = 'T3'
tienda4['Tienda'] = 'T4'

tienda['Facturacion'] = tienda['Precio'] + tienda['Costo de envío']
tienda2['Facturacion'] = tienda2['Precio'] + tienda2['Costo de envío']
tienda3['Facturacion'] = tienda3['Precio'] + tienda3['Costo de envío']
tienda4['Facturacion'] = tienda4['Precio'] + tienda4['Costo de envío']

Tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])
Tiendas.head()
```

#1. Análisis de facturación



### Facturacion total de cada tienda ordenada del mayor a menor

```python
facturacion_por_tienda = Tiendas.groupby('Tienda')['Facturacion'].sum()
facturacion_por_tienda = facturacion_por_tienda / 1000000

lista = facturacion_por_tienda.sort_values(ascending=False)
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(10, 6))
bars = facturacion_por_tienda.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'lightcoral'], width=0.8)
plt.title('Facturación Total por Tienda')
plt.xlabel("Tienda")
plt.ylabel("Facturación")
plt.xticks(rotation=0)

for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
plt.show()
```

### Comparativa por mes de cada tienda para una mejor observacion de ventas

```python
Tiendas['Fecha de Compra'] = pd.to_datetime(Tiendas['Fecha de Compra'])
Tiendas['Mes'] = Tiendas['Fecha de Compra'].dt.month_name()

ordenar_mes = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
venta_mensual = Tiendas.groupby(['Mes', 'Tienda'])['Facturacion'].sum().reset_index()
ventas_tabla = venta_mensual.pivot(index='Tienda', columns='Mes', values='Facturacion')
ventas_tabla
```

Grafico interpretacion visual

```python
ventas_tabla = ventas_tabla[ordenar_mes]

plt.figure(figsize=(12, 6))

for tienda in ventas_tabla.index:
    plt.plot(ventas_tabla.columns, ventas_tabla.loc[tienda], label=tienda, marker='o')

plt.xlabel("Mes")
plt.ylabel("Facturación")
plt.title("Tienda vs Mes")
plt.legend()
plt.grid(True)
```

# 2. Ventas por categoría

### Categoria con el total de ventas por catalogo en orden de mayor ventas

```python
ventas_por_categoria = Tiendas.groupby('Categoría del Producto')['Facturacion'].sum()
lista = ventas_por_categoria.sort_values(ascending=False)
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(12, 6))
plt.bar(ventas_por_categoria.index, ventas_por_categoria.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], width=0.8)
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.title("Ventas por Categoría")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(ventas_por_categoria.values):
    plt.text(i, v, f"{v:.0f}", ha='center')

plt.tight_layout()
plt.show()
```

### Categoria con su total de ventas por tienda ordenado de mayor a menor por tienda

```python
ventas_por_categoria_tienda = Tiendas.groupby(['Tienda', 'Categoría del Producto'])['Facturacion'].sum().reset_index()

lista = ventas_por_categoria_tienda.sort_values(by=['Tienda', 'Facturacion'], ascending=[True, False])
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(12, 6))

for tienda in ventas_por_categoria_tienda['Tienda'].unique():
    tienda_data = ventas_por_categoria_tienda[ventas_por_categoria_tienda['Tienda'] == tienda]
    plt.plot(tienda_data['Categoría del Producto'], tienda_data['Facturacion'], label=tienda, marker='o')

plt.xlabel("Categoría del Producto")
plt.ylabel("Facturación")
plt.title("Facturación por Categoría y Tienda")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


### Promedio por tienda de la calificacion ordenada por la mejor tienda

```python
calificacion_promedio = Tiendas.groupby('Tienda')['Calificación'].mean()

lista = calificacion_promedio.sort_values(ascending=False)
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(12, 6))
plt.bar(calificacion_promedio.index, calificacion_promedio.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], width=0.8)
plt.xlabel("Categoría")
plt.ylabel("Ventas")
plt.title("Ventas por Categoría")
plt.xticks(rotation=0, ha='right')

for i, v in enumerate(calificacion_promedio.values):
    plt.text(i, v, f"{v:.2f}", ha='center')

plt.tight_layout()
plt.show()
```

### Calificacion por mes analisando cual fue la mejor tienda

```python
mejor_mes_tienda = Tiendas.groupby(['Mes', 'Tienda'])['Calificación'].mean().reset_index()

mejor_mes_tienda['Mes'] = pd.Categorical(mejor_mes_tienda['Mes'], categories=ordenar_mes, ordered=True)
mejor_mes_tienda = mejor_mes_tienda.sort_values('Mes')

lista = mejor_mes_tienda.sort_values(by=['Tienda','Calificación'], ascending=[True, False])
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(12, 6))
for tienda in mejor_mes_tienda['Tienda'].unique():
    tienda_data = mejor_mes_tienda[mejor_mes_tienda['Tienda'] == tienda]
    plt.plot(tienda_data['Mes'], tienda_data['Calificación'], label=tienda, marker='o')

plt.xlabel("Mes")
plt.ylabel("Calificación Promedio")
plt.title("Calificación Promedio por Tienda y Mes")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

### Productos mas vendido y menos vendido

```python
producto_mas_vendido = Tiendas.groupby('Producto')['Facturacion'].sum().idxmax()
producto_menos_vendido = Tiendas.groupby('Producto')['Facturacion'].sum().idxmin()

print(f"Producto más vendido: {producto_mas_vendido}")
print(f"Producto menos vendido: {producto_menos_vendido}")
```

Grafico interpretacion visual

```python
productos = [producto_mas_vendido, producto_menos_vendido]
ventas = Tiendas.groupby('Producto')['Facturacion'].sum()[productos]

plt.figure(figsize=(8, 6))
plt.bar(productos, ventas, color=['green', 'red'])
plt.xlabel("Producto")
plt.ylabel("Facturación")
plt.title("Productos Más y Menos Vendidos")
plt.xticks(rotation=45, ha='right')

for i, v in enumerate(ventas):
    plt.text(i, v, f"{v:.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### Producto mas vendido por la tienda

```python
mas_vendidos = Tiendas.groupby(['Mes', 'Tienda', 'Producto'])['Facturacion'].sum().reset_index()
mas_vendidos = mas_vendidos.loc[mas_vendidos.groupby(['Mes', 'Tienda'])['Facturacion'].idxmax()]

mas_vendidos['Mes'] = pd.Categorical(mas_vendidos['Mes'], categories=ordenar_mes, ordered=True)
mas_vendidos = mas_vendidos.sort_values('Mes')
mas_vendidos

lista = mas_vendidos.sort_values(by=['Tienda','Facturacion'], ascending=[True, False])
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(12, 6))

for tienda in mas_vendidos['Tienda'].unique():
    tienda_data = mas_vendidos[mas_vendidos['Tienda'] == tienda]
    plt.plot(tienda_data['Mes'], tienda_data['Producto'], label=tienda, marker='o')

plt.xlabel("Mes")
plt.ylabel("Productos")
plt.title("Producto Mas Vendido por Mes")
plt.legend(title="Tienda")
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Producto menos vendido por la tienda


```python
menos_vendidos = Tiendas.groupby(['Mes', 'Tienda', 'Producto'])['Facturacion'].sum().reset_index()
menos_vendidos = menos_vendidos.loc[menos_vendidos.groupby(['Mes', 'Tienda'])['Facturacion'].idxmin()]

menos_vendidos['Mes'] = pd.Categorical(menos_vendidos['Mes'], categories=ordenar_mes, ordered=False)
menos_vendidos = menos_vendidos.sort_values('Mes')

lista = menos_vendidos.sort_values(by=['Tienda','Facturacion'], ascending=[False, True])
lista
```

Grafico interpretacion visual

```python
plt.figure(figsize=(12, 6))

for tienda in menos_vendidos['Tienda'].unique():
    tienda_data = menos_vendidos[menos_vendidos['Tienda'] == tienda]
    plt.plot(tienda_data['Mes'], tienda_data['Producto'], label=tienda, marker='o')

plt.xlabel("Mes")
plt.ylabel("Productos")
plt.title("Producto Menos Vendido por Mes")
plt.legend(title="Tienda")
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
