```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Todos los datos para usar en los diversos graficos presentados
data = {'Categoría': ['A', 'B', 'C', 'D', 'E', 'F'],
        'ID': [1, 2, 3, 4 , 5, 6],
        'Valor': [10, 20, 15, 25, 5, 28],
        'Valor%': [15, 15, 18, 5, 8, 18],
        'Fecha': ['01/08/2024', '02/08/2024', '03/08/2024', '04/08/2024', '05/08/2024', '06/08/2024'],
        'Subcategoría': ['A1', 'B1', 'C2', 'D1', 'E1', 'F1']}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
sns.barplot(x='Categoría', y='Valor', data=df, palette='pastel')
plt.title('Gráfico de Barras')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='Categoría', y='Valor', data=df, marker='o')
plt.title('Gráfico de Líneas')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Categoría', y='Valor', data=df, hue='Categoría')
plt.title('Gráfico de Dispersión')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Valor'], bins=30, kde=True)
plt.title('Histograma')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Categoría', y='Valor', data=df)
plt.title('Diagrama de Caja')
plt.show()
```

```python
data = np.random.rand(10, 12)
plt.figure(figsize=(10, 5))
sns.heatmap(data, cmap='viridis')
plt.title('Mapa de Calor')
plt.show()
```

```python
df_sorted = df.sort_values(by='ID', ascending=True)


# # Genera el valor en porcentaje
# df_sorted['Generar %'] = df_sorted['Valor'].cumsum() / df_sorted['Valor'].sum() * 100

# Se crea el grafico
fig, ax1 = plt.subplots(figsize=(10, 5))

sns.barplot(x='Categoría', y='Valor', data=df_sorted, palette='pastel', ax=ax1)

ax1.set_xlabel('Categoría')
ax1.set_ylabel('Valor', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Se crea el pareto
ax2 = ax1.twinx()
ax2.plot(df_sorted['Categoría'], df_sorted['Valor%'], color='red', marker='o', linestyle='-')
ax2.set_ylabel('Porcentaje Acumulado (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Linea donde esta el minimo
ax1.axhline(y=30, xmin= 30, color='gray', linestyle='--')

plt.title('Gráfico de Pareto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.pie(df['Valor'], labels=df['Categoría'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Gráfico de Pastel')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Fecha', y='Valor', data=df, hue='Categoría', estimator='sum', ci=None)
plt.fill_between(df['Fecha'], df['Valor'], alpha=0.3)
plt.title('Gráfico de Área')
plt.show()
```

```python
df_pivot = df.pivot_table(index='Categoría', columns='Subcategoría', values='Valor', aggfunc='sum')
df_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gráfico de Barras Apiladas')
plt.show()
```
